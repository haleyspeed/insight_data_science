### Function Definitions

```python
def get_timing (abf, ch):
    if ch == 0:
        abf.setSweep(0, channel = 0)
    elif ch == 2:
        abf.setSweep(0, channel = 1)
    
    i = 0
    step_start = np.nan
    step_end = np.nan
    step_duration = np.nan
    step_amp = np.nan
    for step in abf.sweepC:
        if step != 0.0:
            if np.isnan(step_start) == True:
                step_start = i 
                step_amp = step
            else:
                step_end = i 
        i = i + 1
    
    step_duration = round(step_end - step_start) / sf
    baseline_end = step_start-(20 * sf)
    baseline_start = baseline_end - (100 * sf)
    coupling_end = step_end - (10 * sf)
    coupling_start = coupling_end - (100 * sf) 
    trace_end = step_end + (100 * sf)
    timing = {'trace_start': baseline_start, 'trace_end': trace_end , 'step_duration': step_duration, 'baseline_start':baseline_start, 
              'baseline_end': baseline_end, 'coupling_start': coupling_start , 'coupling_end': coupling_end, 'istim': step_amp}
    #fig, ax = plt.subplots()
    #ax.plot(abf.sweepC)
    return timing

def filter_sweep (abf, sweep, ch, bad_sweeps, timing):
    abf.setSweep (sweepNumber = sweep, channel = ch)
    baseline = np.mean(abf.sweepY [timing['baseline_start']:timing['baseline_end']])
    baseline_sd = np.std(abf.sweepY [timing['baseline_start']:timing['baseline_end']])
    coupling = np.mean(abf.sweepY [timing['coupling_start']:timing['coupling_end']])
    coupling_sd = np.std(abf.sweepY [timing['coupling_start']:timing['coupling_end']])
    rebound = max(abf.sweepY [timing['coupling_end']:timing['trace_end']])
    base_peak = max(abf.sweepY [timing['baseline_start']:timing['baseline_end']])
    base_min = min(abf.sweepY [timing['baseline_start']:timing['baseline_end']])
    coupling_peak = max(abf.sweepY [timing['coupling_start']:timing['coupling_end']])
    trace_peak =  np.max(abf.sweepY [timing['trace_start']:timing['coupling_end']])
    trace_peak_sd =  np.std(abf.sweepY [timing['trace_start']:timing['coupling_end']])
    
    # Find bad traces
    if baseline_sd > abs(baseline * 0.1):
        bad_sweeps.append(sweep)
        print ("Sweep # " + str(sweep) + " removed for wonky baseline")
    elif coupling_sd > abs(coupling * 4) or trace_peak > baseline + 3:
        bad_sweeps.append(sweep)
        print ("Sweep # " + str(sweep) + " removed for wonky coupling region")
    elif baseline > cutoff:
        bad_sweeps.append(sweep)
        print ("Sweep # " + str(sweep) + " removed for Vm > " + str(cutoff))
    #elif trace_peak > baseline + 3 :
        #bad_sweeps.append(sweep)
        #print ("Sweep # " + str(sweep) + " removed for a spike in the trace")
    return bad_sweeps
    
def filter_sweeps (abf, timing):
    display (HTML('<h3> Filtered Sweeps </h3>'))
    bad_sweeps = [0]
    sweeps = abf.sweepList
    for sweep in sweeps:
        #Get X and Y data for this sweep on Ch1 
        for ch in [0,2]:
            bad_sweeps = filter_sweep (abf, sweep, ch, bad_sweeps, timing)
    # Remove bad traces     
    sweeps = [sweep for sweep in sweeps if sweep not in bad_sweeps ]
    display (HTML('<br><br>'))
    return sweeps

def get_trace (abf, sweeps, start, end):
    df0 = pd.DataFrame()
    df2 = pd.DataFrame ()
    
    for sweep in sweeps:
        abf.setSweep (sweepNumber = sweep, channel = 0)  
        df0[str(sweep)] = abf.sweepY [start:end]
        
        abf.setSweep (sweepNumber = sweep, channel = 2)
        df2[str(sweep)] = abf.sweepY [start:end]
            
    df0['avg'] = df0.mean(axis = 1)
    df2['avg'] = df2.mean(axis = 1)
    df0['x'] = abf.sweepX [start:end]
    df2['x'] = abf.sweepX [start:end]
    return df0, df2

# Plot the individual traces
def plot_traces (df_stim, df_rec, timing):
    fig, ax = plt.subplots (2,1, figsize = (8,8), dpi = 100)
    display (HTML('<h3> Trace Raw Data </h3>'))
    display(HTML(df_stim.head().to_html() + '<br>'))
    display (HTML('<h3> Plots </h3>'))
    
    # Individual traces
    #for trace1 in df_stim:
    ax[0].plot(df_stim.x, df_stim.loc[:, df_stim.columns != 'x'], alpha = 0.5)
    ax[1].plot(df_rec.x, df_rec.loc[:, df_rec.columns != 'x'], alpha = 0.5)
    
    # Average traces
    ax[0].plot(df_stim.x, df_stim.avg, lw = 4, color = 'k')
    ax[1].plot(df_rec.x, df_rec.avg, lw = 4, color = 'k')
    
    # Annotate stim params
    baseline_start = timing['baseline_start'] - timing['trace_start']
    baseline_end = timing['baseline_end'] - timing['trace_start']
    coupling_start = timing['coupling_start'] - timing['trace_start']
    coupling_end = timing['coupling_end'] - timing['trace_start']
    base_x = df_stim.x[baseline_start: baseline_end]
    coupling_x = df_stim.x[coupling_start: coupling_end]
    ystim_max = max(df_stim.loc[:, df_stim.columns != 'x'].max())
    ystim_min = min(df_stim.loc[:, df_stim.columns != 'x'].min())
    yrec_max = max(df_rec.loc[:, df_rec.columns != 'x'].max())
    yrec_min = min(df_rec.loc[:, df_rec.columns != 'x'].min())
    
    # Lines
    ax[0].plot(base_x, [ystim_max + 10] * len (base_x), color = "k")
    ax[0].plot(coupling_x, [ystim_max + 10] * len (coupling_x), color = "k")
    
    # Labels
    ax[0].annotate('Baseline', xy = (df_stim.x[baseline_start], ystim_max + 25), color = 'k')
    ax[0].annotate('Coupling', xy = (df_stim.x[coupling_start], ystim_max + 25), color = 'k')
    
    # Prettify
    ax[0].spines['bottom'].set_visible(False)
    ax[0].spines['top'].set_visible(False)
    ax[0].spines['right'].set_visible(False)
    ax[1].spines['top'].set_visible(False)
    ax[1].spines['right'].set_visible(False)
    ax[0].set_xticks([])
    
    # Labels
    ax[1].set_xlabel ('Time (s)')
    ax[0].set_ylabel ('Amplitude (mV)')
    ax[1].set_ylabel ('Amplitude (mV)')
    
    # Set y_limits
    print(ystim_min-100)
    ax[0].set_ylim(ystim_min - 25, ystim_max + 25)
    
    
    
    plt.tight_layout()
    
def get_rin (df):
    df.rin1 = df.vstim1/df.istim1 * 1000
    df.rin2 = df.vstim2/df.istim2 * 1000
    df.rin1_nmda = df.vstim1_nmda/df.istim1_nmda * 1000
    df.rin2_nmda = df.vstim2_nmda/df.istim2_nmda * 1000
    return df
    
def get_cc (df):
    # For cell 1 to cell 2
    df.cc1 = np.where(df.vstim1 != 0, (df.vrec1 / df.vstim1 * 100), df.cc1)
    df.cc1 = np.where(df.vrec1 > 0, 0, df.cc1)
    df.cc1_nmda = np.where(df.istim1_nmda != 0, (df.vrec1_nmda / df.vstim1_nmda * 100), df.cc1_nmda)
    df.cc1_nmda = np.where(df.vrec1_nmda > 0, 0, df.cc1_nmda)
    df.cc1_pchange = np.where(df.cc1 != 0, round(100 * ((df.cc1_nmda - df.cc1)/df.cc1), 2), df.cc1)
    
    # For Cell 2 to Cell 1
    df.cc2 = np.where(df.vstim2 != 0, (df.vrec2 / df.vstim2 * 100), df.cc2)
    df.cc2 = np.where(df.vrec2 > 0, 0, df.cc2)
    df.cc2_nmda = np.where(df.istim2_nmda != 0, (df.vrec2_nmda / df.vstim2_nmda * 100), df.cc2_nmda)
    df.cc2_nmda = np.where(df.vrec2_nmda > 0, 0, df.cc2_nmda)
    df.cc2_pchange = np.where(df.cc2 != 0, round(100 * ((df.cc2_nmda - df.cc2)/df.cc2), 2),df.cc2)
    return df

def get_gj (df):
    # For cell 1 to cell 2
    df.xfer1 = np.where(df.istim1 != 0, (df.vrec1 / df.istim1 * 1000), df.xfer1)
    df.xfer1_nmda = np.where(df.istim1_nmda != 0, (df.vrec1_nmda / df.istim1_nmda * 1000), df.xfer1_nmda)
    df.gj1 = 1/(((df.rin1 * df.rin2 - (df.xfer1 * df.xfer1))/df.xfer1)) * 1000000
    df.gj1_nmda = 1/(((df.rin1_nmda * df.rin2_nmda - (df.xfer1_nmda * df.xfer1_nmda))/df.xfer1_nmda)) * 1000000
    gj1_change = df.gj1_nmda - df.gj1
    df.gj1_pchange = round(100 * (gj1_change / df.gj1), 2)
    
    # For cell 2 to cell 1
    df.xfer2 = np.where(df.istim2 != 0, (df.vrec2 / df.istim2 * 1000), df.xfer2)
    df.xfer2_nmda = np.where(df.istim2_nmda != 0, (df.vrec2_nmda / df.istim2_nmda * 1000), df.xfer2_nmda)
    df.gj2 = 1/(((df.rin2 * df.rin1 - (df.xfer2 * df.xfer2))/df.xfer2)) * 1000000
    df.gj2_nmda = 1/(((df.rin2_nmda * df.rin1_nmda - (df.xfer2_nmda * df.xfer2_nmda))/df.xfer2_nmda)) * 1000000
    gj2_change = df.gj2_nmda - df.gj2
    df.gj2_pchange = round(100 * (gj2_change / df.gj2), 2)
    return df

def get_gin (df):
    df.gin1 = np.where(df.rin1 != 0, ((1 / df.rin1) * 1000), df.gin1)
    df.gin1_nmda = np.where(df.rin1_nmda != 0, ((1 / df.rin1_nmda) * 1000), df.gin1_nmda) # in ns
    df.gin2 = np.where(df.rin2 != 0, ((1 / df.rin2) * 1000), df.gin2)
    df.gin2_nmda = np.where(df.rin2_nmda != 0, ((1 / df.rin2_nmda) * 1000), df.gin2_nmda) # in ns
    return df

def get_bins (df_bin):
    splitter = 10
    df = df_bin.copy()
    for index, row in df.iterrows():
        if row['distance'] >= 0 and row['distance'] < 10:
            df.loc[index:index:, 'bins'] = 5
        elif row['distance'] >= 10 and row['distance'] < 20:
            df.loc[index:index:, 'bins'] = 15
        elif row['distance'] >= 20 and row['distance'] <= 30:
            df.loc[index:index:, 'bins'] = 25
    return df

def get_shrinkage(df_temp):
    df = df_temp.copy()
    for index, row in df.iterrows():
        if row['vstim1'] != 0:
            df.loc[index:index:, 'vstim1_pchange'] = 100*((row['vstim1_nmda'] - row['vstim1'])/row['vstim1'])
        if row['vrec1'] != 0:
            df.loc[index:index:, 'vrec1_pchange'] = 100*((row['vrec1_nmda'] - row['vrec1'])/row['vrec1'])
        if row['vstim2'] != 0:
            df.loc[index:index:, 'vstim2_pchange'] = 100*((row['vstim2_nmda'] - row['vstim2'])/row['vstim2'])
        if row['vrec2'] != 0:
            df.loc[index:index:, 'vrec2_pchange'] = 100*((row['vrec2_nmda'] - row['vrec2'])/row['vrec2'])
    return df
```

## Control Cell 1 to Cell 2

```python
# Format file names
ctr_out1 = ctr_file.replace('.abf','') + '_1-2.csv'

# trace_start, trace_end, step_duration, baseline_start, baseline_end, coupling_start, coupling_end, istim
ctr_timing1 = get_timing(ctr_abf, 0) 

# baseline, baseline_sd, coupling, coupling_sd, rebound
ctr_sweeps = filter_sweeps(ctr_abf, ctr_timing1)
ctr_sweeps1 = [sweep for sweep in ctr_sweeps if sweep % 2 == 0  and sweep not in ctr_exclude_sweeps] # Even sweeps cell 1 -> 2 (python sweep # = clampfit sweep # -1)
ctr_sweeps2 = [sweep for sweep in ctr_sweeps if sweep % 2 != 0  and sweep not in ctr_exclude_sweeps] #  Odd sweeps 2-> 1

# Get step traces
df_ctr0, df_ctr2  = get_trace (ctr_abf, ctr_sweeps1 , ctr_timing1['trace_start'], ctr_timing1['trace_end'])
plot_traces (df_ctr0, df_ctr2, ctr_timing1) # Stim, Rec, Timing

# Get results from averaged data
baseline_start = ctr_timing1['baseline_start']- ctr_timing1['trace_start']
baseline_end = ctr_timing1['baseline_end']- ctr_timing1['trace_start']
coupling_start = ctr_timing1['coupling_start']- ctr_timing1['trace_start']
coupling_end = ctr_timing1['coupling_end']- ctr_timing1['trace_start']

df_sum.vm1 =  [np.mean(df_ctr0.avg[baseline_start:baseline_end])]
df_sum.vm2 = [np.mean(df_ctr2.avg[baseline_start:baseline_end])]
df_sum.vrec1 = [abs(df_sum.vm2[0]) - abs(np.mean (df_ctr2.avg[coupling_start:coupling_end]))  ]
df_sum.vstim1 = [abs(df_sum.vm1[0]) - abs(np.mean (df_ctr0.avg[coupling_start:coupling_end])) ]
df_sum.istim1 = [ctr_timing1['istim']] 

```
